# Q 1] find different payment method and no of transaction and quantity sold?


```python
query = """
    SELECT payment_method,
           SUM(quantity) AS quantity_sold,
           COUNT(unit_price) AS noof_transaction
    FROM walmates
    GROUP BY payment_method;
"""
df = pd.read_sql(query, engine)


fig, ax1 = plt.subplots(figsize=(6, 7))

bars = ax1.bar(df['payment_method'], df['quantity_sold'], color='b', alpha=0.6, label='Quantity Sold')


for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, round(yval, 2), ha='center', color='blue')


ax2 = ax1.twinx()  
line, = ax2.plot(df['payment_method'], df['noof_transaction'], color='r', marker='o', label='No of Transactions')


for i, txt in enumerate(df['noof_transaction']):
    ax2.text(df['payment_method'][i], txt + 0.1, round(txt, 2), ha='center', color='red')


ax1.set_xlabel('Payment Method')
ax1.set_ylabel('Quantity Sold', color='b')
ax2.set_ylabel('Number of Transactions', color='r')
plt.title('Payment Method vs Quantity Sold and Transactions')

# Show legend
ax1.legend(loc='upper right', bbox_to_anchor=(1, 0.85))
ax2.legend(loc='upper right')


plt.xticks(rotation=45)

plt.show()
```

# Q]3] identify busiest day for each branch based on no of transaction?

```python
query = """
SELECT 
    "Branch",
    to_char(date, 'Day') AS day,
    COUNT(*) AS no_of_transactions,
    rank() OVER (PARTITION BY "Branch" ORDER BY COUNT(*)) AS ranked
FROM 
    walmates
GROUP BY 
    "Branch", day
ORDER BY 
    no_of_transactions
"""
df = pd.read_sql(query,engine)

df_top = df[df['ranked'] == 1]

plt.figure(figsize=(10, 19))
bars = plt.barh(df_top['Branch'], df_top['no_of_transactions'])

for bar, day in zip(bars, df_top['day']):
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2, day.strip(), ha='left', va='center')

plt.xlabel('No. of Transactions')
plt.ylabel('Branch')
plt.title('Busiest Day for Each Branch')
plt.tight_layout()

plt.show()
```

# what is average,minimum and maximum rating foe category in each city ?

```python
query = """
SELECT 
    category, 
    "City", 
    AVG(rating) AS avg_rating, 
    MIN(rating) AS min_rating, 
    MAX(rating) AS max_rating
FROM 
    walmates
GROUP BY 
    category, "City";
"""
df = pd.read_sql(query, engine)

top_cities = df.groupby('City').agg({'avg_rating': 'max'}).reset_index()
top_cities = top_cities.sort_values('avg_rating', ascending=False).head(5)

df_top_cities = df[df['City'].isin(top_cities['City'])]

plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

plt.subplot(1, 3, 1)
sns.barplot(data=df_top_cities, x="category", y="avg_rating", hue="City")
plt.title("Average Rating by Category and City")
plt.xticks(rotation=45)

plt.subplot(1, 3, 2)
sns.barplot(data=df_top_cities, x="category", y="min_rating", hue="City")
plt.title("Minimum Rating by Category and City")
plt.xticks(rotation=45)

plt.subplot(1, 3, 3)
sns.barplot(data=df_top_cities, x="category", y="max_rating", hue="City")
plt.title("Maximum Rating by Category and City")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# how many transaction occurs each shift morning,afternoon,evening?

```python
query = """
SELECT
    CASE
        WHEN EXTRACT(HOUR FROM TO_TIMESTAMP(time, 'HH24:MI:SS')) BETWEEN 6 AND 11 THEN 'Morning'
        WHEN EXTRACT(HOUR FROM TO_TIMESTAMP(time, 'HH24:MI:SS')) BETWEEN 12 AND 17 THEN 'Afternoon'
        WHEN EXTRACT(HOUR FROM TO_TIMESTAMP(time, 'HH24:MI:SS')) BETWEEN 18 AND 23 THEN 'Evening'
        ELSE 'Night'
    END AS shift,
    COUNT(*) AS number_of_transactions
FROM 
    walmates
GROUP BY
    shift;
"""
df = pd.read_sql(query,engine)
df.plot(kind='bar', x='shift', y='number_of_transactions', legend=False)
plt.xlabel('Shift')
plt.ylabel('Number of Transactions')
plt.title('Number of Transactions by Shift')
for i in range(len(df)):
    plt.text(i, df['number_of_transactions'][i] + 0.5, df['number_of_transactions'][i], ha='center')
plt.show()
```

# which branches experience  largest decrease in revenue comapare to previous year  by year revenue?

```python
query = """
WITH yearly_revenue AS (
    SELECT 
        "Branch",
        EXTRACT(YEAR FROM date) AS year,
        SUM(total_price) AS revenue
    FROM walmates
    GROUP BY "Branch", EXTRACT(YEAR FROM date)
)
SELECT 
    current."Branch",
    current.year,
    current.revenue - previous.revenue AS revenue_change
FROM yearly_revenue current
LEFT JOIN yearly_revenue previous
    ON current."Branch" = previous."Branch" 
    AND current.year = previous.year + 1
ORDER BY revenue_change ASC;
"""
df = pd.read_sql(query,engine)
top_branches = df.head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_branches['Branch'], top_branches['revenue_change'], color='red')
plt.xlabel('Revenue Change')
plt.ylabel('Branch')
plt.title('Branches with the Largest Decrease in Revenue Year over Year')
for index, value in enumerate(top_branches['revenue_change']):
    plt.text(value, index, str(value))
plt.show()
```
